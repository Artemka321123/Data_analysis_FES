```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent=random_ua)
for i in regions:
    location = geolocator.geocode(i)
    print(location.address)
    print((location.latitude, location.longitude))
```

```python
Moscow = (55.7505412, 37.6174782)
```

```python
dict_distance = dict()
from geopy.distance import geodesic 
for i in regions:
    location = geolocator.geocode(i)
    dict_distance[i] = geodesic(Moscow, (location.latitude, location.longitude)).km
```

```python
df_dist = pd.DataFrame.from_dict(dict_distance, orient='index')
```
